# Column transformation

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
data = {
    'liquid': ['tea', 'coffee', 'water','soft_drink',  'beer'],
    'temp': ['hot', 'cold', np.nan, 'cold', 'cold'],
    'price': [25, 35, 15, None, 30],
    'quantity_ml': [100, 200, None, 100, 300],
    'target': [0, 1, 0, 1, 1]
   
}

df = pd.DataFrame(data)
df
```

```python
X = df.drop('target', axis=1)
X
```

```python
num_feature = ['price','quantity_ml']
cat_feature = ['temp','liquid']
```

```python
num_transformer = Pipeline(
    steps = [
        ('imputation', SimpleImputer(strategy='median')),
        ('scaling', StandardScaler())
    ]
)
```

```python
cat_transformer = Pipeline(
    steps = [
        ('imputation', SimpleImputer(strategy='most_frequent')),
        ('encoding', OneHotEncoder(handle_unknown='ignore'))
    ]
)
```

```python
transformer = ColumnTransformer(
    transformers = [
        ('numeric_transformations', num_transformer, num_feature),
        ('categorical_transformations', cat_transformer, cat_feature)
    ]
)
```

```python
X_processed = transformer.fit_transform(X)
X_processed
```
